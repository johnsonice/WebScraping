```python
import os
from os import listdir
from os.path import isfile, join
```

```python
# Convert english punctuation to chinese punctuation.
def convert(enchar):
    if enchar == ',': return '、'
    elif enchar == '.': return '。'
    elif enchar == '!': return '！'
    elif enchar == '?': return '？'
    elif enchar == ';': return '；'
    elif enchar == ':': return '：'
    print("No Chinese conversion found.")
    return enchar



# This method takes the punctuation from an english line and adds it to a chinese line.
def addpunc(enline, cnline):
    enline = list(enline)
    cnline = list(cnline)
    punctuation = [',', '.', '!', '?', ';', ':']
    
    # Add all the punctuation from the english line to list named enpunc, in the order they appear.
    enpunc = []
    for c in enline:
        if c in punctuation:
            enpunc.append(c)
    
    # If the english line has no punctuation, just return.
    if len(enpunc) == 0:
        cnline = "".join(cnline)
        print("No punctuation")
        return cnline
    
    # If there is only one punctuation mark in the english statement. 
    elif len(enpunc) == 1:
        
        # If it comes at the end, then put it at the end.
        if enline[len(enline)-2] in punctuation:
            cnline.insert(len(cnline)-1, convert(enpunc[0]))
            i = 0
            #Fill in the spaces with commas.
            while (i < len(cnline)):
                if cnline[i] == ' ':
                    cnline[i] = '、'
                    i = i + 1
                i = i + 1
            print("Only one, at end.")
            
        # Otherwise, just put it in the first space.
        else:
            cnline["".join(cnline).find(' ')] = convert(enpunc[0])
            print("Only one, not at end.")
        
        cnline = "".join(cnline)
        return cnline
    
    # Other cases: add stuff in the middle.
    else:
        # Add all of the punctuation in the middle. Keep spaces after puncutation since each space is actually two spaces.
        enpunc_counter = 0
        i = 0
        while (i < len(cnline)):
            if cnline[i] == ' ':
                cnline[i] = convert(enpunc[enpunc_counter])
                if not enpunc_counter == len(enpunc) - 1:
                    enpunc_counter = enpunc_counter + 1
                i = i + 1
            i = i + 1
        
        # Add the ending punctuation.
        cnline.insert(len(cnline)-1, convert(enpunc[len(enpunc)-1]))
        cnline = "".join(cnline)
        return cnline
```

```python
enfolder = os.path.expanduser("~/Documents/AI_Work/data_final/srt/entest/")
cnfolder = os.path.expanduser("~/Documents/AI_Work/data_final/srt/cntest/")

# Lists of filenames.
enfiles = [f for f in listdir(enfolder) if isfile(join(enfolder, f))]
cnfiles = [f for f in listdir(cnfolder) if isfile(join(cnfolder, f))]

# Loop through all of the files.
for x in range(len(enfiles)):
    
    # Read in files.
    cnfilelocation = "~/Documents/AI_Work/data_final/srt/cntest/" + cnfiles[x]
    enfilelocation = "~/Documents/AI_Work/data_final/srt/entest/" + enfiles[x]
    cnfile = os.path.expanduser(cnfilelocation)
    enfile = os.path.expanduser(enfilelocation)
    cnline = ''
    enline = ''
    
    #Create the writefile.
    writefile = os.path.expanduser("~/Documents/AI_Work/cn_punctuation_test/" + cnfiles[x] + "_WithPunctuation.txt")
    writer = open(writefile, "w")

    with open(cnfile, 'r') as fcn, open(enfile, 'r') as fen:
        cnline = fcn.readline()
        enline = fen.readline()

        while (not cnline == '') & (not enline == ''):
            print('Chinese: ', cnline)
            print('English: ', enline)
            cn_withpunc = addpunc(enline, cnline)
            print('Chinese with punctuation: ', cn_withpunc)
            writer.write(cn_withpunc)
            print()
            print()
            cnline = fcn.readline()
            enline = fen.readline()
        
    writer.close()


print()
print("Done.")
```
