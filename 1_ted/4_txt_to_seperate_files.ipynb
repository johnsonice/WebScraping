### Convert one file to two files 

```python
import os 
from os import path
from itertools import compress
import pandas as pd
```

```python
##############################
#### define some functions ###
##############################
def check(f):
    filename, file_extension = os.path.splitext(f)
    if file_extension == '.txt':
        return True
    else:
        return False
    
def check_length(e,c):
    if len(e)==len(c):
        return True
    else:
        return False
```

```python
os.getcwd()
```

```python
data_folder = 'data/cn_tw/result_folder/'
result_folder_cn = 'data/cn_tw/result_folder2/cn/'
result_folder_en = 'data/cn_tw/result_folder2/en/'
problem_folder = 'data/cn_tw/problem/'
```

```python
## check if extension is .srt
txts = [data_folder + f for f in os.listdir(data_folder) if check(f)]
len(txts)
```

```python
txts[0]
```

```python
doc_list ={'title':[],'status':[]}
txt_list = txts
for index,txt in enumerate(txt_list):
    print (len(txt_list),index+1)
    with open(txt, 'r') as f:
        myFile = [line.strip() for line in f]
    myFile = [x for x in myFile if x != '']
    sp = [s.split('|') for s in myFile]
    en =[ele[0].strip() for ele in sp]
    cn = [ele[1].strip() for ele in sp]
    fname = txt.split('/')[-1].split('.')[-2]
    doc_list['title'].append(fname)
    if check_length(en,cn):
        with open(result_folder_cn + str(index) + '_' + fname + '.txt', mode='wt', encoding='utf-8') as myfile:
            myfile.write('\n'.join(cn))
        with open(result_folder_en + str(index) + '_' + fname + '.txt', mode='wt', encoding='utf-8') as myfile:
            myfile.write('\n'.join(en))            
        doc_list['status'].append('Success')
    else:
        doc_list['status'].append('Failed')
```

```python
df = pd.DataFrame(doc_list)
df_problem=df[df.status=='Failed']
```
